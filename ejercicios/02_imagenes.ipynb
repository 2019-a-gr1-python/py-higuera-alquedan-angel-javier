```python
import numpy as np
from scipy import ndimage
from scipy import misc
import matplotlib.pyplot as plt
```

```python
mapache = misc.face()
```

```python
recortar_mapache = mapache[10:110, 10:110]
recortar_mapache.shape
plt.title("Mapache recorte")
plt.imshow(recortar_mapache)
```

```python
plt.title("Mapache negativo")
mapache2= 255-mapache
plt.imshow(mapache2)
```

```python
plt.title("Mapache ruidoso")
mapache_random=mapache.flatten().reshape(768, 1024, 3)
for row in mapache_random:
    for pixel in row:
        pixel[0]=np.random.randint(low=1, high=255)
#         pixel[1]=np.random.randint(low=1, high=255)
        pixel[2]=np.random.randint(low=1, high=255)
plt.imshow(mapache_random)
plt.show()
```

```python
recortar_mapache = mapache[250:-250, 250:-250]
recortar_mapache.shape
plt.title("Mapache recorte")
plt.imshow(recortar_mapache)
```

```python
mapache.shape
recortar_mapache_1 = mapache[0:384, 0:512]
recortar_mapache_2 = mapache[385:768, 0:512]
recortar_mapache_3 = mapache[0:384, 513:1024]
recortar_mapache_4 = mapache[385:768, 513:1024]
parte_1=np.concatenate((recortar_mapache_4, recortar_mapache_2), axis=1)
parte_2=np.concatenate((recortar_mapache_3, recortar_mapache_1), axis=1)
mapache_recorte=np.concatenate((parte_1, parte_2))
recortar_mapache.shape
plt.title("Mapache recorte")
plt.imshow(mapache_recorte)
```

```python
plt.title("Mapache ruidoso")
mapache_random = np.flipud(mapache)
plt.imshow(mapache_random)
plt.show()
```

```python
copia_mapache = np.copy(mapache)
mapache_borroso = ndimage.gaussian_filter(copia_mapache, sigma=10)
plt.imshow(mapache_borroso)
plt.show()
```
